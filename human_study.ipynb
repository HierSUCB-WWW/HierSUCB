```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
```

```python
ppl=[31,38,16,4]
round_20=np.array([[6,3,18,24],
[18,23,1,9],
[9,32,2,8],
[9,33,2,7],
[10,29,3,9]])
round_50=np.array([[8,34,2,7],
[20,21,1,9],
[8,32,2,9],
[11,33,2,5],
[9,28,4,10]])
percent_r20=np.array([np.average(round_20[:,0],axis=0),np.average(round_20[:,1],axis=0),np.average(round_20[:,2],axis=0),np.average(round_20[:,3],axis=0)])*100/51
div_r20=np.std(round_20*100/51,axis=0)
percent_r50=np.array([np.average(round_50[:,0],axis=0),np.average(round_50[:,1],axis=0),np.average(round_50[:,2],axis=0),np.average(round_50[:,3],axis=0)])*100/51
div_r50=np.std(round_50*100/51,axis=0)
names=['C2UCB','Hier-SUCB','Both','Neither']
big_names=['Like\nEither','Dislike\nBoth']
# print(percent_r20,percent_r50)
C2UCB=[percent_r20[0],percent_r50[0]]
HSUCB=[percent_r20[1],percent_r50[1]]
BOTH=[percent_r20[2],percent_r50[2]]
NEITHER=[percent_r20[3],percent_r50[3]]
print(C2UCB,HSUCB,BOTH,NEITHER)
```

```python
rounds = ['Round 20', 'Round 50']
plt.figure(figsize=(6, 6))
plt.fill_between(rounds, C2UCB, color='#c80f7d', label='C2UCB', alpha=0.8, edgecolor='none')
plt.fill_between(rounds, np.array(C2UCB) + np.array(BOTH), C2UCB, color='#2b2f8e', label='Both', alpha=0.8, edgecolor='none')
plt.fill_between(rounds, np.array(C2UCB) + np.array(BOTH) + np.array(HSUCB), 
                 np.array(C2UCB) + np.array(BOTH), color='#00a6e6', label='HSUCB', alpha=0.8, edgecolor='none')
plt.fill_between(rounds, np.array(C2UCB) + np.array(BOTH) + np.array(HSUCB) + np.array(NEITHER), 
                 np.array(C2UCB) + np.array(BOTH) + np.array(HSUCB), color='grey', label='Neither', alpha=0.8, edgecolor='none')

# Remove axes, ticks, and labels
plt.axis('off')

# Save as SVG
plt.savefig('stacked_area_chart.svg', format='svg', bbox_inches='tight', pad_inches=0)
plt.close()
```

```python
def percentage_formatter(x,pos):
    return f'{x:.0f}%'
bar_height = 0.35
bar_positions50 = range(len(percent_r20))
bar_positions20 = [pos + bar_height for pos in bar_positions50]

plt.figure(figsize=(7.5,5),dpi=300)
plt.subplots_adjust(left=0.20,right=0.95,bottom=0.18,top=0.98)
plt.barh(bar_positions20,percent_r20,height=bar_height,label='Round 20',color='mediumaquamarine')
plt.barh(bar_positions50,percent_r50,height=bar_height,label='Round 50',color='coral')
ax = plt.gca()  # Get the current axis
ax.set_xticks(np.linspace(0,100,5))
ax.xaxis.set_major_formatter(FuncFormatter(percentage_formatter))
plt.yticks([pos + bar_height / 2 for pos in bar_positions50], big_names,fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Type of Preferred Visualization',fontsize=24)
plt.ylabel('Percentage of Users',fontsize=24)
ax.tick_params(axis='y', which='both', bottom=False, top=False)
plt.legend(fontsize=20)
plt.savefig('ustudy.pdf')
```
